4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?
EmpID	Name	Department
E101	Raju	IT
E102	Riya	HR
E103	Roshni	Finance
E104	Sejal	Sales

Ans. From the given data, the Department of Employee ID E103 is → Finance

 Formula using VLOOKUP:
=VLOOKUP("E103", A2:C5, 3, FALSE)
 Explanation:
"E103" → value we want to search
A2:C5 → table range (EmpID to Department)
3 → column number where Department is located
FALSE → exact match
 Which LOOKUP function is used and why?

We use VLOOKUP because:

Data is arranged vertically (in columns)
We search for EmpID in the first column
We want to return a value (Department) from a column to the right

 If data were horizontal, we would use HLOOKUP instead.

 | EmpID | Name   | Department |
| ----- | ------ | ---------- |
| E101  | Raju   | IT         |
| E102  | Riya   | HR         |
| E103  | Roshni | Finance    |
| E104  | Sejal  | Sales      |

 So, E103 → Finance

 5. A company has set the following commission structure for its sales team Sales Amount 0 20000 50000 80000 Commission Rate 5% 10% 15% 20% If a salesperson achieves sales of ₹45,000, use the VLOOKUP function to determine the commission rate applicable.

 Ans:
 Formula using VLOOKUP:
=VLOOKUP(45000, A2:B5, 2, TRUE)

 Explanation:
45000 → sales amount
A2:B5 → table (Sales Amount & Commission Rate)
2 → return value from 2nd column (Commission Rate)
TRUE → approximate match (finds closest lower value)

| Sales Amount | Commission |
| ------------ | ---------- |
| 0            | 5%         |
| 20000        | 10%        |
| 50000        | 15%        |
| 80000        | 20%        |

Answer:
₹45,000 falls between 20,000 and 50,000
So applicable commission rate = 10%
 Why TRUE (Approximate Match)?
Because we are working with ranges/slabs, not exact values
VLOOKUP finds the nearest lower value and returns its commission

 So, final answer: 10% commission
